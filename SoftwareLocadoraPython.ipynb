```python
def portfolio():

    print("=========================================")
    print("[0] Chevrolet Tracker - R$ 120,00/dia")
    print("[1] Chevrolet Onix - R$ 90,00/dia")
    print("[2] Chevrolet Spin - R$ 150,00/dia")
    print("[3] Hyunday HB20 - R$ 85,00/dia")
    print("[4] Hyunday Tucson - R$ 120,00/dia")
    print("[5] Fiat Uno - R$ 70,00/dia")
    print("[6] Fiat Mobi - R$ 70,00/dia")
    print("[7] Fiat Pulse - R$ 130,00/dia")
    print("=========================================")



def alugar_carro(x):

    quantidade_dias = int(input("Escolha a quantidade de dias : "))

    while quantidade_dias<0:

        quantidade_dias = int(input("Dias Inválido. Escolha uma quantidade de dias válido."))

    preco = {0:120,1:90,2:150,3:85,4:120,5:70,6:70,7:130}

    print("Você escolheu alugar {} por {} dias, totalizando R${},00 !".format(catalogo[x],quantidade_dias,(preco[x]*quantidade_dias)))

    lista_alugados.append(x)



def devolver_carro():

    print("Qual carro você deseja devolver?")

    print("Segue a lista de carros alugados. Qual você deseja devolver?")

    if len(lista_alugados)==0:

        print("Não há carros a serem devolvidos")

    else:

        for carro in lista_alugados:

            print("[{}] - {}".format(carro,catalogo[carro]))

        carro_devolvido = int(input())

        print("O carro {} foi devolvido com sucesso".format(catalogo[carro_devolvido]))

        lista_alugados.remove(carro_devolvido)



def menu():

    print("=======================================================")
    print("Bem vindo a locadora de carros!")
    print("O que deseja fazer?\n0 - Mostrar Portfólio | 1 - Alugar carro | 2 - Devolver carro")
    print("=======================================================")

    opcao = int(input("Digite a opção desejada : "))

    while opcao <0 or opcao>2:

        opcao=int(input("Erro! Digite uma opção válida : "))

    ## Mostrar portfólio
    if opcao == 0:

        portfolio()

        escolha = int(input("0 - Voltar | 1 - Sair  "))

        while escolha<0 or escolha>1:

            escolha = int(input("Erro. Escolha uma opção válida"))

        if escolha ==1:
          print("Até mais. Foi um prazer ter você aqui.")

        if escolha ==0:

            menu()


    ## Alugar carro
    elif opcao == 1:

        portfolio()

        codigo_carro = int(input("Escolha o código do carro : "))

        while codigo_carro<0 or codigo_carro>7:

            codigo_carro = int(input("Erro. Escolha um código válido : "))

        alugar_carro(codigo_carro)

        escolha = int(input("0 - Voltar | 1 - Sair  "))

        while escolha<0 or escolha>1:

            escolha = int(input("Erro. Escolha uma opção válida"))

        if escolha ==1:
          print("Até mais. Foi um prazer ter você aqui.")

        if escolha ==0:

            menu()

    else:

        devolver_carro()

        escolha = int(input("0 - Voltar | 1 - Sair  "))

        while escolha<0 or escolha>1:

            escolha = int(input("Erro. Escolha uma opção válida"))

        if escolha ==1:
          print("Até mais. Foi um prazer ter você aqui.")

        if escolha ==0:

            menu()


catalogo = {0:"Chevrolet Tracker",1:"Chevrolet Onix",2:"Chevrolet Spin",3:"Hyunday HB20",4:"Hyunday Tucson",5:"Fiat Uno",6:"Fiat Mobi",7:"Fiat Pulse"}

lista_alugados=[]

menu()
```
